```python
import bs4
import string
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = urlopen('https://www.newegg.com/p/pl?d=mobile')
print('Gagan Chordia - 19BCE0788')
```

```python
page = url.read()
url.close()
```

```python
soup = BeautifulSoup(page, 'html.parser')
products = soup.findAll("div", {"class": "item-container"})
```

```python
file = "prod.csv"
f = open(file, "w")
headers = ("Product Name,Price,Discount,Image\n")
f.write(headers)
```

```python
for prod in products:
    name = prod.findAll("a", {"class": "item-title"})
    prod_name = name[0].text
    price = prod.findAll("li", {"class": "price-current"})
    prod_price = price[0].text
    disc = prod.findAll("li", {"class": "price-save"})
    prod_disc = disc[0].text
    img = prod.img["src"]
    f.write(prod_name.replace(",", "|") + "," + str(prod_price) + "," + str(prod_disc) + "," + img + "\n")
    print('Product Name: ' + prod_name + "\nProduct Price: " + prod_price + "\nDiscount: " + prod_disc + "\nImage: " + img)
    
f.close()
```
