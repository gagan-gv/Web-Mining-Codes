```python
def remove_stopwords(para, stopword_list):
    words_in_para = para.split()
    res = ""
    
    for word in words_in_para:
        word = word.lower()
        if word not in stopword_list:
            new_word = ""
            for ch in word:
                if ch not in ['.', ',']:
                    new_word += ch
                    
            res += (new_word + " ")
    print("Code by Gagan Chordia - 19BCE0788")
    print('Paragraph before removal of stop words: ')
    print(para, '\n')
    print('Paragraph after removal of stop words: ')
    print(res)
```

```python
para = "This is a demo paragraph to demonstrate the removal of stop words, given in the question."

stopwords_list = ['.',',','a','they','the','his','so','and','were','from','that','of','in','only','with','to']

remove_stopwords(para, stopwords_list)
```

```python
#2a

def tokenize_sentence(sentence):
    special_character_list = ['!','"','#','$','%','&','(',')','*','+','/',':',';','<','=','>','@','[','\\',']','^','`','{','|','}','~','\t',',','.']
    
    words = sentence.split()
    res = []
    
    for word in words:
        word = word.lower()
        
        new_word = ""
        
        for ch in word:
            if ch not in special_character_list:
                new_word += ch
                
        res.append(new_word)

    #print("Tokenized data: ")
    return res
```

```python
sentence = "Hey, my name is Gagan Chordia."

print(tokenize_sentence(sentence))
```

```python
#2b

def tokenize_multiple_sentences(sentences):
    sentences_list = sentences.split('.')
    tokens = []
    for sentence in sentences_list:
        tokens.append(tokenize_sentence(sentence))
        
    return tokens
```

```python
sentences = "Hey, my name is Gagan Chordia. I'm currently studying in VIT."

print(tokenize_multiple_sentences(sentences))
```
