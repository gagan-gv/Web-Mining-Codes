```python
import matplotlib.pyplot as plt
import networkx as nx
print('Gagan Chordia - 19BCE0788')
```

```python
#q1
graph1 = {
    "a": ['b', 'e', 'f'],
    "b": ['a', 'c'],
    "c": ['b', 'd', 'f'],
    "d": ['c', 'g'],
    "e": ['a', 'i'],
    "f": ['a', 'c', 'g', 'j'],
    "g": ['d', 'f', 'h', 'j', 'k'],
    "h": ['g', 'l', 'o'],
    "i": ['e', 'j', 'm', 'n'],
    "j": ['f', 'g', 'i', 'n', 'o'],
    "k": ['g', 'o'],
    "l": ['h', 'p'],
    "m": ['i'],
    "n": ['i', 'j', 'o'],
    "o": ['h', 'j', 'k', 'p'],
    "p": ['l', 'o']
}
```

```python
G = nx.DiGraph(graph1)
nx.draw_networkx(G, with_labels=True)
```

```python
def cent_degree(g):
    cent = dict()
    
    print('N = ', len(g))
    
    for i in g:
        cent[i] = len(g[i]) / (len(g) - 1)
    
    return cent
```

```python
print('Centrality Measure - Degree')
print(cent_degree(graph1))
```

```python
def cent_closeness(g):    
    print('N = ', len(g))
    _sum = 0
    
    for i in g:
        _sum += len(g[i])
        
    return (len(g) - 1) / _sum
```

```python
print('Centrality Measure - Closeness')
print(cent_closeness(graph1))
```

```python
print('Centrality Measure - Betweenness')
print(nx.betweenness_centrality(G))
```

```python
def pres_degree(g, G):
    return dict((v, len(G.in_edges(v)) / (len(g) - 1)) for v in G.nodes())
```

```python
print('Prestige Measure - Degree')
print(pres_degree(graph1, G))
```

```python
def pres_proximity(G):
    proximity = dict()

    temp_prox = 0
    n = 0
    for dest in G.nodes:
        temp_prox = 0
        n = 0
        for src in G.nodes:
            if (nx.has_path(G,src,dest) == True):
                temp_prox = temp_prox + nx.shortest_path_length(G,source = src,target = dest)
                n = n + 1
        if temp_prox == 0:
            proximity[dest] = 0
        else:
            proximity[dest] = temp_prox/(n - 1)
    return proximity
```

```python
print('Prestige Measure - Proximity')
print(pres_proximity(G))
```
