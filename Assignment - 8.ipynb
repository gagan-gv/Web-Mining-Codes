## Gagan Chordia



```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import operator
import random as rd
print('Gagan Chordia - 19BCE0788')
```

```python
graph = {
    1: [2, 3],
    2: [],
    3: [1, 2, 5],
    4: [5, 6],
    5: [4, 6],
    6: [4]
}
```

```python
G = nx.DiGraph(graph)
nx.draw_networkx(G, with_labels=True)
plt.show()
```

```python
rank_dict={}

starting_node = 1

for ii in range(1, 7):
    rank_dict[ii] = 1
    
rank_dict[starting_node] = rank_dict[starting_node] + 1

for ii in range(0, 7):
    list_n=list(G.neighbors(starting_node))
    
    if(len(list_n) == 0):
        starting_node = rd.randint(0, 6)
        rank_dict[starting_node] = rank_dict[starting_node] + 1
    else:
        starting_node = rd.choice(list_n)
        rank_dict[starting_node] = rank_dict[starting_node] + 1
        
    for ii in range(1, 7):
        rank_dict[ii] /= 6
    
rank_dict_sorted=sorted(rank_dict.items(),key=lambda v:(v[1],v[0]),reverse=True)

print('Page Rank: ')
for ii in rank_dict_sorted:
    print(ii[0], end=" ")
```
