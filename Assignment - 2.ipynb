### Gagan Chordia - 19BCE0788



```python
#importing required libraries

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
#1a

sentence = "Hey there, I am Gagan Chordia."

stop_words = set(stopwords.words('english'))
tokenized = word_tokenize(sentence)

new_sentence = [w for w in tokenized if not w.lower() in stop_words]

print('After stop words removal:\n', new_sentence)
```

```python
lem = WordNetLemmatizer()

temp = []

for w in new_sentence:
    w = lem.lemmatize(w)
    if w not in temp:
        temp.append(w)
    
print('After lemmatization:\n', temp)
```

```python
stem = PorterStemmer()

new_sentence = ''

for w in temp:
    w = stem.stem(w)
    if w not in new_sentence:
        new_sentence += (w + ' ')
        
print('After stemming:\n', new_sentence)
```

```python
res = []
special_character_list = ['!','"','#','$','%','&','(',')','*','+','/',':',';','<','=','>','@','[','\\',']','^','`','{','|','}','~','\t',',','.']

temp = new_sentence.split()

for w in temp:
    word = ""
    for ch in w:
        if ch not in special_character_list:
            word += ch
    res.append(word)

print('Final tokens:\n', res)
```
